```python
import numpy as np
import data
from copy import deepcopy
```

# k = 3

```python
k=3
n=200
dimension=2
mu_list = np.array([ [2,6], [8,10], [8,2] ])
sigma_list = np.array([ [[1,0],[0,3]], [[3,0],[0,2]], [[3,0],[0,3]] ])
X = data.generate_data(k, n, dimension, mu_list, sigma_list)
real_lable = deepcopy(X[:, -1])
print(mu_list)
data.show(X, mu_list, title='Real Data Distribution')
```

### k-means

```python
X, center, iterations = data.kmeans(X, k, epsilon=1)
print(center)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center, title='epsilon=1, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
X, center, iterations = data.kmeans(X, k, epsilon=1e-5)
print(center)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center, title='epsilon=1e-5, iterations=' + str(iterations) + 'accuracy=' + str(accuracy))
X, center, iterations = data.kmeans(X, k, epsilon=1e-7)
print(center)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center, title='epsilon=1e-7, iterations=' + str(iterations) + 'accuracy=' + str(accuracy))
```

```python
k=5
n=200
dimension=2
mu_list = np.array([ [2,2], [2,8], [5,5], [8,2], [8,8] ])
sigma_list = np.array([ [[1,0],[0,2]], [[3,0],[0,1]], [[2,0],[0,2]], [[2,0],[0,3]], [[1,0],[0,1]] ])
X = data.generate_data(k, n, dimension, mu_list, sigma_list)
real_lable = deepcopy(X[:, -1])
print(mu_list)
data.show(X, mu_list, title='Real Data Distribution')
```

```python
X, center, iterations = data.kmeans(X, k, epsilon=1)
print(center)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center, title='epsilon=1, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
X, center, iterations = data.kmeans(X, k, epsilon=1e-5)
print(center)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center, title='epsilon=1e-5, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
X, center, iterations = data.kmeans(X, k, epsilon=1e-7)
print(center)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center, title='epsilon=1e-7, iterations=' + str(iterations) + 'accuracy=' + str(accuracy))
```

### GMM

```python
X, iterations,mu_list = data.gmm(X, k, 1)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center=mu_list,title='epsilon=1, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
X, iterations,mu_list= data.gmm(X, k, epsilon=1e-5)
accuracy = data.accuracy(real_lable, X[:, -1], k)
data.show(X, center=mu_list,title='epsilon=1e-5, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
```

# UCI 数据集

```python
X = data.uci_iris()
real_lable = deepcopy(X[:, -1])
k = 3
# k-means
print('\nK-means')
X, center, iterations = data.kmeans(X, k, epsilon=1)
accuracy = data.accuracy(real_lable, X[:, -1], k)
print(center)
data.show(X, center, title='epsilon=1, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
print('epsilon=1, iterations=' + str(iterations) + ', accuracy=' + str(accuracy) + '\n')

X, center, iterations = data.kmeans(X, k, epsilon=1e-5)
accuracy = data.accuracy(real_lable, X[:, -1], k)
print(center)
data.show(X, center, title='epsilon=1e-5, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
print('epsilon=1e-5, iterations=' + str(iterations) + ', accuracy=' + str(accuracy) + '\n')

X, center, iterations = data.kmeans(X, k, epsilon=1e-10)
accuracy = data.accuracy(real_lable, X[:, -1], k)
print(center)
data.show(X, center, title='epsilon=1e-10, iterations=' + str(iterations) + ', accuracy=' + str(accuracy))
print('epsilon=1e-10, iterations=' + str(iterations) + ', accuracy=' + str(accuracy) + '\n')

# GMM
print('\nGMM')
```
